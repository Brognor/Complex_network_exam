```python
import time 
import evolution as evo
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
f = open("datatime.txt","w")
```

```python
for i in range (2,2000):
    f = open("datatime.txt","a")
    G=nx.complete_graph(i)
    vec=np.ones(i)
    vec=evo.normalize(vec)
    start=time.time()

    evo.evolution_collection(G,vec,100,norm=False)

    end=time.time()

    f.write(f"{i}\t{round((end-start),3)}\n")
    f.close()
    with open("datatime.txt","r") as f:
        lines=f.readlines()
        x=[float(line.split()[0]) for line in lines]
        y=[float(line.split()[1]) for line in lines]
        fig=plt.figure()
        ax1= fig.add_subplot(111)

        ax1.set_title("Execution time")
        ax1.set_xlabel("Number of nodes")
        ax1.set_ylabel("Time")

        ax1.plot(x,y,c='r', linewidth=1)
        fig.savefig("transition_time.png")
        plt.show()
    
```

```python
f.close()
```
